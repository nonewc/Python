```python
### 교통사고 파일을 열어서 DataFrame으로 생성
## 1. 생성된 DataFrame을 통해서 총 사건수, 사망자 수, 중상자 수, 경상자 수 조회
## 2. 서울 지역 사망자수를 조회하여 총 사망자 수를 조회

import pandas as pd

ca = pd.read_csv('data/교통사고.csv',encoding="utf-8")

city = ca['시도']
ac = ca['사고건수']
dt = ca['사망자수']
si = ca['중상자수']
ij = ca['경상자수']

print('총 사건수 : ',sum(ac))
print('사망자 수 : ',sum(dt))
print('중상자 수 : ',sum(si))
print('경상자 수 : ',sum(ij))

seoul = 0
for idx in range(len(ca)) :
    if city[idx] == '서울':
        seoul += dt[idx]

print("서울지역 사망자 수 : ",seoul)

### 한국도로공사_휴게소 주차장 현황
## 1. 노선별 주차장의 합계를 딕셔너리로 데이터를 조회 생성
## 2. 각 본부별 주차장 수를 딕셔너리 데이터로 조회 생성

park = pd.read_csv('data/휴게소.csv',encoding="euc-kr")

lines = {}
for i in range(len(park)):
    lines[park['노선'][i]] = lines.get(park['노선'][i],0) + park['합계'][i]
print(lines)

## 딕셔너리 내포
key = ['1','2','3','4']
value = [1,2,3,4]
test_dic = { key[i]:value[i] for i in range(4)}
print(test_dic)

parks = {}
for i in range(len(park)):
    parks[park['본부'][i]] = parks.get(park['본부'][i],0) + park['합계'][i]
    
print(parks)
```

```python
jumsu = {
'홍길동' : [50,60,50],
'이순신' : [60,60,60],
'강감찬' : [100,90,90],
'유관순' : [80,80,90],
'강환석' : [95,95,95]
}

## 4개의 Series로
# 이름(name_S), 각 리스트 0번은 국어(kor_S), 1번은 (eng_S),
# 2번은 수학(math_S)
##

na = [key for key in jumsu]
name_S = pd.Series(na)

kor = [jumsu[key][0] for key in na]
kor_S = pd.Series(kor)

eng = [jumsu[key][1] for key in na]
eng_S = pd.Series(eng)

math = [jumsu[key][2] for key in na]
math_S = pd.Series(math)

tot = [sum(jumsu[key]) for key in na]
tot_S = pd.Series(tot)

# total = kor_S + eng_S + math_S

mean_S = round(tot_S/3,2)

## DataFrame 만들기
score = pd.DataFrame({'이름':name_S,'국어':kor_S,'영어':eng_S,'수학':math_S})

score['합계'] = tot_S
score['평균'] = mean_S

score

```
