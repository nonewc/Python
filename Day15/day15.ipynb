```python
## 특정 디렉터리 내에 있는 텍스트 자료 수정
import os

# 1. 텍스트 디렉터리 경로 지정
print(os.getcwd())  # 현재 기본 경로 확인
txt_data = 'data/txt_data/'

# 2. 텍스트 디렉터리 목록 반환
sub_dir = os.listdir(txt_data)
print(sub_dir)

# 3. 각 디렉터리의 텍스트 자료 수집 함수
def textPro(sub_dir):    # first,second
    first_txt = []       # first 디렉터리 텍스트 저장
    second_txt = []      # second 디렉터리 텍스트 저장
    
    # 디렉터리 구성
    for sdir in sub_dir:
        dirname = txt_data + sdir
        file_list = os.listdir(dirname)
        
        # 파일 구성
        for fname in file_list:
            file_path = dirname + '/' + fname
            
            # file 선택
            if os.path.isfile(file_path):
                try:
                    # 텍스트 자료 수집
                    file = open(file_path,'r')
                    if sdir == 'first'
                        first_txt.append(file.read())
                    else:
                        second_txt.append(file.read())
                except Exception as e :
                    print("예외 발생 : ",e)
                finally:
                    file.close()
                    
    return first_txt, second_txt # 텍스트 자료 반환
                    
        
```

```python
### 특수한 파일 처리
# CSV(Comma Separated Value) : 열단위 자료를 처리할 수 있는 대표적인
# 파일로 CSV, Excel -> Cell 단위 자료를 저장하여 처리하는 파일

# pandas 패키지 import
import pandas as pd
import os

os.getcwd()

score = pd.read_csv('data/score.csv')
print(score.info()) # 파일 정보
print(score.head()) # 컬럼명 포함 앞 5개 행 출력

# 칼럼 추출
kor = score.kor     # 객체.컬럼명
eng = score['eng']  # 객체['컬럼명']
mat = score['mat']
dept = score['dept']

print(sum(kor))

# 각 과목의 최고점 알아오기
print('max kor : ',max(kor))
print('max eng : ',max(eng))
print('max math : ',max(mat))

# 각 과목의 최하점 알아오기
print('min kor : ',min(kor))
print('min eng : ',min(eng))
print('min math : ',min(mat))

# 각 과목의 평균
from statistics import mean
print("국어 점수 평균 : ",round(mean(kor),2))
print("영어 점수 평균 : ",round(mean(eng),2))      
print("수학 점수 평균 : ",round(mean(mat),2))

# dept 빈도수
dept_count = {}

for key in dept:
    dept_count[key] = dept_count.get(key,0) + 1
print(dept_count)
    
```

```python
### JSON 파일
# JSON(JavaScript Object Notation)은 네트워크 표준으로 사용되고 있는
# 파일 양식.(XML...)
# JSON파일의 특징은 "키-값" 쌍으로 자료를 표현
# 표현은 자바스크립트 구문형식을 따르지만, 프로그램 언어나 플랫폼에
# 독립적으로 동작할 수 있음

# 파이썬은 JSON 파일을 처리할 수 있는 모듈을 기본적으로 제공하고 있음
# 인코딩 - 파이썬 객체(dict or lsit)를 -> JSON 문자열로 변환
# 디코딩 - JSON 문자열("키-값"을 -> 파이썬 객체로 변환

# 인코딩 : json.dumps(객체) -> json문자열로 변경
# 디코딩 : json.loads(json문자열) -> 파이썬 객체

```

```python
import json

## json 인코딩
user = {'id':1234, 'name':'홍길동'} # 파이썬 딕셔너리
print(type(user))
print(user['name'])

jsonString = json.dumps(user,ensure_ascii=False)
# ensuer_ascii 옵션에 False = 아스키코드를 사용하지 않아요
print(type(jsonString))
print(jsonString)

#디코딩
pyObj = json.loads(jsonString)
print(type(pyObj))
print(pyObj)
print(pyObj['name'])
```
